```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
insurance = pd.read_csv('/Users/vaibhavrangan/Downloads/Stat_303-2/Datasets/insurance.csv')
insurance.head()
```

```python
# create scatterplot of bmi vs charges
sns.scatterplot(x = "bmi", y = "charges", data = insurance)
```

```python
x = insurance[["bmi"]]
y = insurance["charges"]

from sklearn.model_selection import train_test_split

slr = LinearRegression()
slr.fit(x, y)
print(slr.coef_, slr.intercept_)
```

```python
# statsmodels.formula.api

import statsmodels.formula.api as smf
import statsmodels.api as sm

ols_object = smf.ols(formula = "charges ~ bmi", data = insurance)
model = ols_object.fit()
model.summary()

# same intercept as sklearn trained model
```
